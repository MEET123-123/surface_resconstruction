```python
# 工具函数定义
########################################################
def getA_B(data):
    ret=[]
    for i in range(len(data)-1):
        data1,data2=data[i],data[i+1]
        ur1,ndi1=data1[0],data1[1]
        ur2,ndi2=data2[0],data2[1]
        A=np.dot(ndi2,np.linalg.inv(ndi1))
        B=np.dot(ur2,np.linalg.inv(ur1))
        tmp=[A,B]
        ret.append(tmp)
    return ret

def get_tip_points(ndidata):
    # T->TX
    Pret=np.zeros((4,1))
    Pret[0][0], Pret[1][0], Pret[2][0],Pret[3][0]=round(ndidata[0][3],3),round(ndidata[1][3],3),round(ndidata[2][3],3),1
    return Pret

def get_transfer_ret(Rce, Tce):
    # RX(3,3),TX(3,1)->T
    ret = np.zeros((4, 4))
    for i in range(3):
        for j in range(3):
            ret[i][j] = Rce[i][j]
        ret[i][3] = Tce[i][0]
    ret[3][3] = 1
    return ret

def  get_transfer_ret_(Rx,Tx):
    # RX(3,3),TX(3,)->T
    ret=np.zeros((4,4))
    for i in range(3):
        for j in range(3):
            ret[i][j]=Rx[i][j]
        ret[i][3]=Tx[i]
    ret[3][3]=1
    return ret

def get_rx_tx(data):
    # T->RX,TX
    rx = np.zeros((3, 3))
    tx = np.zeros((3, 1))
    for i in range(3):
        for j in range(3):
            rx[i][j] = data[i][j]
        tx[i] = data[i][3]
    return rx, tx

def transpose(matrix):  # 转置
    ret = np.zeros((len(matrix[0]), len(matrix)))
    for i in range(len(matrix[0])):
        for j in range(len(matrix)):
            ret[i][j] = matrix[j][i]
    return ret

 
def EulerAngles2RotationMatrix(theta, format='degree'):
    """
    Calculates Rotation Matrix given euler angles.
    :param theta: 1-by-3 list [rx, ry, rz] angle in degree
    :return:
    RPY角，是ZYX欧拉角，依次 绕定轴XYZ转动[rx, ry, rz]
    """
    if format is 'degree':
        theta = [i * math.pi / 180.0 for i in theta]
 
    R_x = np.array([[1, 0, 0],
                    [0, math.cos(theta[0]), -math.sin(theta[0])],
                    [0, math.sin(theta[0]), math.cos(theta[0])]
                    ])
 
    R_y = np.array([[math.cos(theta[1]), 0, math.sin(theta[1])],
                    [0, 1, 0],
                    [-math.sin(theta[1]), 0, math.cos(theta[1])]
                    ])
 
    R_z = np.array([[math.cos(theta[2]), -math.sin(theta[2]), 0],
                    [math.sin(theta[2]), math.cos(theta[2]), 0],
                    [0, 0, 1]
                    ])
    R = np.dot(R_z, np.dot(R_y, R_x))
    return R
 
print(EulerAngles2RotationMatrix([120,0,0]))

# Checks if a matrix is a valid rotation matrix.
def isRotationMatrix(R) :
    Rt = np.transpose(R)
    shouldBeIdentity = np.dot(Rt, R)
    I = np.identity(3, dtype = R.dtype)
    n = np.linalg.norm(I - shouldBeIdentity)
    return n < 1e-6


# Calculates rotation matrix to euler angles
# The result is the same as MATLAB except the order
# of the euler angles ( x and z are swapped ).
def RotationMatrixToEulerAngles(R) :
 
    assert(isRotationMatrix(R))
     
    sy = math.sqrt(R[0,0] * R[0,0] +  R[1,0] * R[1,0])
     
    singular = sy < 1e-6
 
    if  not singular :
        x = math.atan2(R[2,1] , R[2,2])
        y = math.atan2(-R[2,0], sy)
        z = math.atan2(R[1,0], R[0,0])
    else :
        print("y轴90度-死锁状态")
#         x = math.atan2(-R[1,2], R[1,1])
#         y = math.atan2(-R[2,0], sy)
#         z = 0
    angle=np.array([x, y, z])/math.pi*180
    return angle
 
```

```python
def readNow():
    dic={'MessageSize': 'H', 'Reserved_1': '3H', 'DigitalInputs': 'Q', 'DigitalOutputs': 'Q', 'RobotMode': 'Q',          'TimeStamp': 'Q',
         'Reserved_2': 'Q',  'TestValue': 'Q',   'Reserved_3': 'd',    'SpeedScaling': 'd',   'LinearMomentumNorm': 'd', 'VMain': 'd', 
         'VRobot': 'd',      'IRobot': 'd',      'Reserved_4': 'd',    'Reserved_5': 'd',     'ToolAcceleroMeter': '3d', 'ElbowPosition': '3d', 
         'ElbowVelocity': '3d', 'QTarget': '6d', 'QDTarget': '6d',     'QDDTarget': '6d',     'ITarget': '6d',           'MTarget': '6d',
         'QActual': '6d',    'QDActual': '6d',   'IActual': '6d',      'ActualTCPForce': '6d','ToolVectorActual': '6d',  'TCPSpeedActual': '6d',
         'TCPForce': '6d',   'ToolVectorTarget': '6d','ToolSpeedTarget': '6d',
         'MotorTemperatures': '6d', 'JointModes': '6d','VActual': '6d','HandType': '4c',      'User': 'c',               'Tool': 'c', 
         'RunQueuedCmd': 'c','PauseCmdFlag': 'c','VelocityRatio': 'c', 'AccelerationRatio': 'c','JerkRatio': 'c',        'XYZVelocityRatio': 'c',
         'RVelocityRatio': 'c','XYZAccelerationRatio': 'c','RAccelerationRatio': 'c','XYZJerkRatio': 'c','RJerkRatio': 'c','BrakeStatus': 'c',
         'EnableStatus': 'c','DragStatus': 'c', 'RunningStatus': 'c', 'ErrorStatus': 'c',    'JogStatusCR': 'c',         'CRRobotType': 'c',
         'DragButtonSignal': 'c','EnableButtonSignal': 'c','RecordButtonSignal': 'c','ReappearButtonSignals': 'c',       'JawButtonSignal': 'c',
         'SixForceOnline': 'c','SixForceOnline': 'c','Reserve[82]':'82c','MActual[6]':'6d',   'load':'d',                'CenterX':'d',
         'CenterY':'d',      'CenterZ':'d',      'User[6]':'6d',       'Tool[6]':'6d',        'TraceIndex':'d',          'SixForceValue[6]':'6d',
         'TargetQuaternion[4]':'4d','ActualQuaternion[4]':'4d','Reserve3[24]':'24s'}
    HOST = "192.168.1.6"    # The remote host
    PORT = 30004        # The same port as used by the server
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect((HOST, PORT))
    
    names=[]
    data=s.recv(1440)
    for i,key in enumerate(dic):
        fmtsize=struct.calcsize(dic[key])#计算fmt长度
        data1,data=data[0:fmtsize],data[fmtsize:]#分割当前信息
        fmt=dic[key]
        names.append(struct.unpack(fmt, data1))#解码
        dic[key]=dic[key],struct.unpack(fmt, data1)
    RobotPos=dic["ToolVectorActual"][1]#实际位置
    isStop=dic["QDActual"][1]#实际关节速度
    rx=RobotPos[3:]
    rx=EulerAngles2RotationMatrix(rx)
    tx_=RobotPos[:3]
    ret=get_transfer_ret_(rx,tx_)
    tx=np.zeros((4,1))
    tx[0][0],tx[1][0],tx[2][0]=round(tx_[0],3),round(tx_[1],3),round(tx_[2],3)
    tx[3][0]=1
    return RobotPos,ret,tx,isStop[0]  # 机械臂6自由度数据、旋转矩阵、位移量、当前速度
```

```python
def eye_hand_calibration():#获取ndi和ur数据
    #######################初始化NDI##############################
    SETTINGS = {
        "tracker type": "polaris",
#         "romfiles": ["./NDI_Rom/8700339qjx.rom"]
        "romfiles": ["NDI_Rom/8700340_2022_11_21.rom"]
    }
    TRACKER = NDITracker(SETTINGS)
    TRACKER.start_tracking()
    
    #######################初始化机械臂##############################
    HOST = "192.168.1.6"
    PORT = 30003  
    s = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    s.connect((HOST, PORT))
    data_row=1  # 数据标志位
    while (1):
        data = []
        #print(i)
        str1 = "MovJ("
        str2 = ",Tool=2)"
        f = open("calibration_robot_points.txt", "r")
        #print("f:",f)
        ur_list,ndi_list=[],[]
        data_index=0
        for line in f:
            #print("data_row:",data_row)
            #print("line:", line)
            #tmpdata=[]
            time.sleep(1)
            line = line.split("[")
            line = line[1]
            line = line.split("]")
            line = line[0]
            strL = str1 + line + str2
            strL = bytes(strL, encoding="ascii")
            s.send(strL)
            data_row+=1
            #print("strL:", strL)
            time.sleep(1.5)
            #######################运动停止检测##############################
            i=0
            while(1):
                RobotPos,flage_transfer,flage_tx,isStop = readNow()  # 机械臂向量、旋转矩阵、旋转矩阵tx、关节速度
                #print(isStop)
                if abs(isStop)<0.03:
                    flage_Position=flage_tx
                    i+=1
                if abs(isStop)>0.5:#表明仍在运动，重新计数
                    i=0
                if i==5:
                    break
            time.sleep(1.5)
            raw_track_index=0
            ndi_data_isRight=1
            while(1):
                port_handles, timestamps, framenumbers, tracking, quality = TRACKER.get_frame()
                ret = tracking[0]
                if np.isnan(ret[0][0]) == True:
                    raw_track_index+=1
                # print("ret[3][0]:",ret[3][0])
                # 连续监测15个空值则停止
                if raw_track_index==15 and np.isnan(ret[0][0]) == True:
                    ndi_data_isRight=0
                    NDIret=-1
                    NDI_position=np.zeros((4,4))
                    break
                NDIret=np.zeros((4,4))
                if np.isnan(ret[0][0]) == False:  #np.isnan(T_tipB2W[0][0])
                    #print("data_index:",data_index)
                    #print(ret)
                    for j in range(10):
                        port_handles, timestamps, framenumbers, tracking, quality = TRACKER.get_frame()
                        NDIret +=tracking[0]
                    NDIret=NDIret/10
                    NDI_position=get_tip_points(NDIret)
                    break
            data_index+=1
            if np.isnan(NDI_position[0][0])==False and NDI_position[0][0]!=0 :
                tmpdata=[flage_transfer,NDIret]
                data.append(tmpdata)
                flage_Position=transpose(flage_Position)
                NDI_position=transpose(NDI_position)
#                 print("NDI_position：",NDI_position)
                if len(ur_list)==0:
                    ur_list=flage_Position
                    ndi_list=NDI_position
                else:
                    ur_list=np.concatenate((ur_list,flage_Position),axis=0)  # n*4矩阵
                    ndi_list = np.concatenate((ndi_list, NDI_position), axis=0)  # n*4矩阵
                #i+=1
        f.close()
        break
    TRACKER.stop_tracking()
    TRACKER.close()
    ur_list = transpose(ur_list)  # 转为4*n
    ndi_list = transpose(ndi_list)  # 转为4*n
    ndi_ur=np.dot(ndi_list,np.linalg.pinv(ur_list))
    ndi_ur[3][0],ndi_ur[3][1],ndi_ur[3][2]=0,0,0 # 最后一行直接置为0
    
    rx_ret=np.zeros((3,1))
    for i in range(len(data)):
        T_tip2B=data[i][0]  # 机械臂下针尖
        T_w2tip=np.linalg.inv(data[i][1])
        T_tip2w=np.dot(ndi_ur,T_tip2B)
        T_tipB2W=np.dot(T_w2tip,T_tip2w)
        rx,tx=get_rx_tx(T_tipB2W)
        rx=cv2.Rodrigues(rx)[0]  # 旋转矩阵转为旋转向量，做旋转向量的均值
        rx_ret+=rx
    rx_ret=rx_ret/len(data)
    rx_ret=cv2.Rodrigues(rx_ret)[0]  # 旋转向量转回旋转矩阵，这一步不涉及机械臂的读取和交互，因此不需要考虑使用欧拉角还是旋转向量
    tx=np.zeros((4,1))
    tx[3][0]=1
    T_tipB2W=get_transfer_ret(rx_ret,tx)
    return ndi_ur,T_tipB2W,ur_list,ndi_list
```
